```python
categorcial_variables = ['job', 'marital', 'education']

for col in categorcial_variables:
    plt.figure(figsize=(20,10))
    plt.subplots_adjust(hspace=0.5)
    #Returns counts of unique values for each outcome for each feature.
    pos_counts = df.loc[df.deposit.values == 'yes', col].value_counts() 
    neg_counts = df.loc[df.deposit.values == 'no', col].value_counts()
    
    all_counts = list(set(list(pos_counts.index) + list(neg_counts.index)))
    
    #Counts of how often each outcome was recorded.
    freq_pos = (df.deposit.values == 'yes').sum()
    freq_neg = (df.deposit.values == 'no').sum()
    
    pos_counts = pos_counts.to_dict()
    neg_counts = neg_counts.to_dict()
    
    all_index = list(all_counts)
    all_counts = [pos_counts.get(k, 0) / freq_pos - neg_counts.get(k, 0) / freq_neg for k in all_counts]

    sns.barplot(all_counts, all_index)
    plt.title(col)
    plt.savefig("plot8.png", bbox_inches='tight')
   
```

```python
def plot_hist(variable):
    plt.figure(figsize=(9,6))
    plt.hist(df[variable], bins=40,color='#cd1076')
    plt.xlabel(variable)
    plt.ylabel("frequency")
    plt.title("{} distrubition with hist".format(variable))
    plt.show()
numericVar = ["age","campaign","duration"]

```

```python
for n in numericVar:
    plot_hist(n)
```
